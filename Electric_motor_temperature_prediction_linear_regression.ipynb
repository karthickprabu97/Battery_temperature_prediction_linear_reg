```python
df2 = df.sample(n=10000, random_state=101)
df2.head(2)
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(df2.corr(), cmap="BuGn", vmin=0, vmax=1, annot=True);
```

```python
df2.columns
```

```python
X = df2.drop(['profile_id','stator_winding','torque','pm'],axis =1)
y = df2['pm']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
X.columns
```

```python
pd.DataFrame(data = lm.coef_, index = X.columns, columns = ['Coefficients'])
```

```python
y_predicted = lm.predict(X_test_scaled)
```

```python
plt.scatter(y_test, y_predicted);
```

```python
residuals = y_test - y_predicted
```

```python
sns.distplot(residuals);
```

```python
from sklearn import metrics

mae = metrics.mean_absolute_error(y_test, y_predicted)
mse = metrics.mean_squared_error(y_test, y_predicted)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_predicted))
R_2 = metrics.r2_score(y_test, y_predicted)

print(f'MAE is: {mae}')
print(f'MSE is: {mse}')
print(f'RMSE is: {rmse}')
print(f'R^2 is: {R_2}')
```
